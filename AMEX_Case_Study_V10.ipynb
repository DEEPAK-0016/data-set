```python
df = train1[cols1]
```

```python
def data_processor():
    features_to_select = ['campaign_id','product','webpage_id', 'product_category_1','user_group_id', 'gender', 'age_level', 'user_depth',
       'city_development_index', 'var_1']

    print('Use StandardScaler to process the column data')
    scaler = StandardScaler()
    df[df.columns[:-1].tolist()] = scaler.fit_transform(df[df.columns[:-1].tolist()])
    # print(df.head(5))
    X = df[df.columns[:-1].tolist()]
    X = X[features_to_select]
    y = df[df.columns[-1]]

    print("Train Test Split ratio is 0.3")
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=0)
    print(f"X_train shape: {X_train.shape}")
    print(f"y_train shape: {y_train.shape}")

    return X_train, X_test, y_train, y_test
```

```python
X_train, X_test, y_train, y_test = data_processor()
```

```python
# logistic Regression

clf = LogisticRegression(C=1e5)
clf.fit(X_train, y_train)
# build model with SMOTE imblearn
smote_pipeline = make_pipeline_imb(SMOTE(random_state=4), clf)

smote_model = smote_pipeline.fit(X_train, y_train)
smote_prediction = smote_model.predict(X_test)
smote_prediction_proba = smote_model.predict_proba(X_test)[:, 1]
```

```python
from sklearn import metrics
metrics.accuracy_score(y_test,smote_prediction)
```

```python
import seaborn as sn
lg_cm = metrics.confusion_matrix(y_test,smote_prediction,[1,0])
lg_cm
```

```python
# Random forest 

clf1 = RandomForestClassifier(max_depth=5, n_estimators=100)
clf1.fit(X_train, y_train)
# build model with SMOTE imblearn
smote_pipeline = make_pipeline_imb(SMOTE(random_state=4), clf1)

smote_model = smote_pipeline.fit(X_train, y_train)
smote_prediction = smote_model.predict(X_test)
smote_prediction_proba = smote_model.predict_proba(X_test)[:, 1]
```

```python
print("F1 score is: {}".format(f1_score(y_test, smote_prediction)))
print("AUC Score is: {}".format(roc_auc_score(y_test,smote_prediction_proba )))
```
